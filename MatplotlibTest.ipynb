```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
np.random.seed(2000)
y = np.random.standard_normal((20,2))
y = y.cumsum(axis=0)
y[:, 1] = y[:, 1] * 5 
```

```python
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 1], 'r', lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple plot with unsymmetry red & blue line')
```

```python
fig, ax1 = plt.subplots() ## Crucial
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=8)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
ax2 = ax1.twinx() ## Crucial
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(211)  ## numrows, numcols, fignum
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
#plt.xlabel('index')
plt.title('A simple plot')
plt.subplot(212) ## numrows, numcols, fignum
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
plt.xlabel('index')
```

```python
plt.figure(figsize=(9, 4))
plt.subplot(121)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title("1st Data set")
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:,0], width=0.5, color='g', label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.title('2nd Data set')
```

```python
from matplotlib.patches import Polygon
def func(x):
    return 0.5 * np.exp(x) + 1

a, b = 0.5, 1.5
x = np.linspace(0,2,100)
y = func(x)

fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)

Ix = np.linspace(a, b, 100)
Iy = func(Ix)
verts = [(a,0)] + list(zip(Ix, Iy)) + [(b,0)]
poly = Polygon(verts, facecolor='0.8', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5 * (a+b), 1, r"$\int_a^b f(x)\mathrm{d}x$", horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
ax.set_xticks((a,b))
ax.set_xticklabels(("$a$", "$b$"))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
plt.grid(True)
```

```python
from pandas_datareader import data as web
start = "2019-11-01"
end = "2020-01-02"

df = web.DataReader('^GDAXI', data_source='yahoo', start = start, end = end)
df = df.reset_index()
#print(df)
#print(df["Date"])
df["Date"] = df["Date"].apply(lambda x: x.toordinal())
print(df.columns)
df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']] ## Crucial
print(df.columns)
#print(df["Date"])
#print(df)
#print(df.to_records(index=False))
#print(type(df.to_records(index=False)))
quotes = [tuple(x) for x in df.to_records(index=False)]
#print(quotes)
#print(type(quotes))
#print(quotes[:10])
```

```python
import mpl_finance as mpf
fig, ax = plt.subplots(figsize=(8,5))
fig.subplots_adjust(bottom=0.2)
mpf.candlestick_ohlc(ax, quotes, width=0.6, colorup='b', colordown='r')
plt.grid(True)
ax.xaxis_date()
ax.autoscale_view()
plt.setp(plt.gca().get_xticklabels(), rotation=30)
print("")
#mpf.candlestick_ohlc?
```

```python
fig, ax = plt.subplots(figsize=(8,5))
mpf.plot_day_summary_ohlc(ax, quotes, colorup='b', colordown='r')
plt.grid(True)
ax.xaxis_date()
plt.title('DAX index')
plt.ylabel("index level")
plt.setp(plt.gca().get_xticklabels(), rotation=30)
print("")
```
